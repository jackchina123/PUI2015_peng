#The range of X line and Y line should be in [20000, 65000].
#The order of race in the plot legend is the order of the pecentage of the races in the U.S. 
#The star is for all the race, the circle is for specific race.
# I choose the distinct color and obvious size for the race points so that the points can be shown clear.
#The color of the points of Hispanic, Black and Asian are carefully selected.

```python
pl.figure(figsize=(10,10))
col={'All':'Green','White':'Blue','Black':'Black','Hispanic':'Red','Asian':'Orange'}

for k in ['All']:
    pl.scatter(allmales[k]['Mean income'][1],allfemales[k]['Mean income'][1],s=250*allmales[k]\
               ['Mean income'][1]/allfemales[k]['Mean income'][1],marker=(5, 1), c=col[k],label=k)
for k in ['White', 'Hispanic', 'Black', 'Asian']:
    pl.scatter(allmales[k]['Mean income'][1],allfemales[k]['Mean income'][1],s=150*allmales[k]\
               ['Mean income'][1]/allfemales[k]['Mean income'][1],c=col[k],label=k)
pl.plot([20000,65000],[20000,65000],'k--')
pl.xlabel('Male Mean Income',fontsize=15)
pl.ylabel('Female Mean Income',fontsize=15)
pl.xlim(20000, 65000)
pl.ylim(20000, 65000)
pl.title("Females VS. Males in Mean Income by Race",fontsize=20)
pl.legend(loc=2,fontsize=15)

```
