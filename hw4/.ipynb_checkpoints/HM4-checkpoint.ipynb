```python
df['ageM'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
df['ageF'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 2)]
```

```python
bins = np.arange(10, 99, 10)
```

```python
df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).plot(kind='bar')
df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).plot(kind='bar')
```

```python
csM = df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).cumsum()
csF = df.ageM.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).cumsum()
print np.abs(csM / csM.max() - csF / csF.max())

print 'x',bins[:-1] + 5
print 'y',csM.max()
pl.plot(bins[:-1] + 5, csM / csM.max(), label = "M")
pl.plot(bins[:-1] + 5, csF / csF.max(), label = "F")
pl.legend()
```

```python
df['daybiker'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['starttime'].dt.hour < 18 )]
df['nightbiker'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['starttime'].dt.hour >=18)]
```

```python
csD = df.daybiker.groupby(pd.cut(df.daybiker, bins)).agg([count_nonzero]).cumsum()
csN = df.nightbiker.groupby(pd.cut(df.nightbiker, bins)).agg([count_nonzero]).cumsum()
print np.abs(csD / csD.max() - csN / csN.max())
```
